```python
# pandas와 matplotlib을 사용하여 타이타닉 데이터셋을 다운로드하고, 남녀 생존 비율을 바차트로 시각화
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

# 한글 폰트 설정 (Windows의 경우 'Malgun Gothic')
matplotlib.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

# 타이타닉 데이터셋 다운로드
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# 남성과 여성의 생존 비율 계산
gender_survived = df.groupby('Sex')['Survived'].mean() * 100

# 바차트로 시각화
plt.figure(figsize=(6,4))
gender_survived.plot(kind='bar', color=['skyblue', 'pink'])
plt.title('타이타닉호 남녀 생존 비율')
plt.ylabel('생존 비율 (%)')
plt.xlabel('성별')
plt.ylim(0, 100)
plt.xticks(rotation=0)
plt.show()
```
